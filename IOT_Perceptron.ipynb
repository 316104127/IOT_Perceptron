```python
"""
Created on Fri Jun 2020
@author: Sidnei Alves de Araújo
Implementação do Perceptron conforme algoritmo dado em aula 
"""


import numpy as np

#passo 0
def le_conj_trein(text_arq):
    arq = open(text_arq, 'r')  #abre o arquivo
    linhas_arq = []  #declara uma matriz para receber as linhas do arquivo
    matriz = [] #declaro uma matriz para receber o conjunto de treinamento
    linhas_arq = arq.readlines() #quebra o arquivo em linhas e armazena em linhas_arq
    arq.close()  #fecha o arquivo
    for i in range(len(linhas_arq)):          
        matriz.append(linhas_arq[i].split())  #aqui separa os valores contidos em cada linha do arquivo 
    #converte os valores lidos em números e coloca no conj de treinamento
    ct = np.zeros( (len(matriz),len(matriz[0])),'float')
    #print(matriz)
    for i in range(ct.shape[0]):
        for j in range(ct.shape[1]):
            ct[i][j] = float(matriz[i][j])
    return ct # retorna uma a matriz no formato ndarray 

#passo 1    
def incializa_pesos(w):
    for i in range(len(w)):
        w[i] = np.random.uniform(-1, 1)
        
    return w

#passo 2    
def calc_s(x,w):
    s = 0
    for i in range(len(x)):
        s = s + x[i] * w[i]  
    return s

#passo 3    
def calc_y(s):
    if s>=0: return 1
    else: return -1

#passo 4    
def calc_erro(d,y):
    return d-y

#passo 5
def atualiza_pesos(wt, x, erro):
    wt1 = wt
    for i in range(len(wt)):
        wt1[i] = wt[i] + TX_AP * erro * x[i]
    return wt1

#Fazer predição para um padrão p
def predicao_rna(p):
    print("======")
    x = np.append([BIAS],p, axis=0)
    print(f"Predição de X{x}")
    s = calc_s(x,w)
    print(f"Predição de S{s}")
    y = calc_y(s)
    print(f"Predição de Y{y}")
    print("======")
    
    return y

#############################################  Passos da RNA Percepetron ######################
#lê o conjunto de treinamento (passo 0)
ct = le_conj_trein('Porta_OR.txt')
num_padroes = ct.shape[0]
num_atrib = ct.shape[1]-1  
print('Conjunto de Treinamento:')
print(ct)
#separa padroes (p) e saídas desejadas (d)
p = ct[:,0:num_atrib]
print(p)
d = ct[:,num_atrib]
print(d)

w = []
#Passo 1 do algoritmo
#Parametros da RNA
BIAS    = 1.0  #Fixa o valor de Bias
TX_AP   = 0.1  #Taxa de Aprendizagem
EPOCAS  = 10000
#cria e inicializa vetor de pesos
w = np.zeros((num_atrib+1),'float')

print("*******************************")
print("teste 1")
w = incializa_pesos(w)
#w = [0.5, 0.7, 0.9, 0.1]
cont = 1
print('Conjunto de pesos iniciais:')
print(w)

while(cont < EPOCAS):
    #print(f"ÉPOCA:{cont}--------")
    erro_tot = 0
    for i in range(num_padroes):
      
      x = np.append([BIAS],p[i], axis=0)
      #print(f"X: {x}")
      #Passo 2 do algoritmo - Para o padrão i Calcula S
      s = calc_s(x,w)
      #print(f"S: {s}")
      #Passo 3 do algoritmo - Calcula y
      y = calc_y(s)
      #print(f"Y: {y}")
      
      #Passo 4 do algoritmo - Calcula erro     
      erro = calc_erro(d[i], y)
      #print(f"ERRO: {erro}")
      
      #Passo 5 do algoritmo - Se erro <> zero, atualiza pesos
       
      if erro !=0:
          #print(w)  
          w = atualiza_pesos(w,x,erro)
          erro_tot = erro_tot + np.abs(erro)
    #verifica critério de parada
    if erro_tot == 0:
        break
    #incrementa contador de épocas
    cont=cont+1
    
    
    

print('Treinamento finalizado com', cont, 'Epocas')
print('Conjunto de pesos finais:')
print(w)    

# Predições teste
print("Predições teste")
padrao = [0.5,	0.7, 0.1] 
pred = predicao_rna(padrao)
print('Valor de saída para o padrao ', padrao, ' = ', pred)




print("*******************************")
print("teste 2")
w = incializa_pesos(w)
#w = [0.9, 0.2, 0.9, 0.4]
cont = 1
print('Conjunto de pesos iniciais:')
print(w)

while(cont < EPOCAS):
    #print(f"ÉPOCA:{cont}--------")
    erro_tot = 0
    for i in range(num_padroes):
      
      x = np.append([BIAS],p[i], axis=0)
      #print(f"X: {x}")
      #Passo 2 do algoritmo - Para o padrão i Calcula S
      s = calc_s(x,w)
      #print(f"S: {s}")
      #Passo 3 do algoritmo - Calcula y
      y = calc_y(s)
      #print(f"Y: {y}")
      
      #Passo 4 do algoritmo - Calcula erro     
      erro = calc_erro(d[i], y)
      #print(f"ERRO: {erro}")
      
      #Passo 5 do algoritmo - Se erro <> zero, atualiza pesos
       
      if erro !=0:
          #print(w)  
          w = atualiza_pesos(w,x,erro)
          erro_tot = erro_tot + np.abs(erro)
    #verifica critério de parada
    if erro_tot == 0:
        break
    #incrementa contador de épocas
    cont=cont+1
    
    
    

print('Treinamento finalizado com', cont, 'Epocas')
print('Conjunto de pesos finais:')
print(w)    

# Predições teste
print("Predições teste")
padrao = [0.6,	0.9, 0.8] 
pred = predicao_rna(padrao)
print('Valor de saída para o padrao ', padrao, ' = ', pred)



print("*******************************")
print("teste 3")
w = incializa_pesos(w)
cont = 1
print('Conjunto de pesos iniciais:')
print(w)

while(cont < EPOCAS):
    
    erro_tot = 0
    for i in range(num_padroes):
      
      x = np.append([BIAS],p[i], axis=0)
      
      #Passo 2 do algoritmo - Para o padrão i Calcula S
      s = calc_s(x,w)
      
      #Passo 3 do algoritmo - Calcula y
      y = calc_y(s)
      
      
      #Passo 4 do algoritmo - Calcula erro     
      erro = calc_erro(d[i], y)
      
      
      #Passo 5 do algoritmo - Se erro <> zero, atualiza pesos
       
      if erro !=0:
           
          w = atualiza_pesos(w,x,erro)
          erro_tot = erro_tot + np.abs(erro)
    #verifica critério de parada
    if erro_tot == 0:
        break
    #incrementa contador de épocas
    cont=cont+1
    
    
    

print('Treinamento finalizado com', cont, 'Epocas')
print('Conjunto de pesos finais:')
print(w)    

# Predições teste
print("Predições teste")
padrao = [0.3,	0.7, 0.2] 
pred = predicao_rna(padrao)
print('Valor de saída para o padrao ', padrao, ' = ', pred) 
```
